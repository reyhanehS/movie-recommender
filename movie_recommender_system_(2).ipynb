##Build the Movie Recommender System

Now we have to convert our values to the form that can be used by our recommender system

```python
from ast import literal_eval
import pandas as pd

features = ["cast", "keywords", "genres"]

for feature in features:
    movies_new[feature] = movies_new[feature].apply(lambda x: literal_eval(x) if pd.notna(x) else [])

movies_new[features].head(10)
```

```python
from ast import literal_eval
def get_list(x):
    if isinstance(x, list):
        names = [i["name"] for i in x]

        if len(names) > 3:
            names = names[:3]

        return names

    return []
```

```python
features = ["cast", "keywords", "genres"]
for feature in features:
    movies_new[feature] = movies_new[feature].apply(get_list)
```

```python
movies_new.head()
```

Now we're gonna transform the above feature instances into lowercase and eliminate all the spaces among them.

```python
def clean_data(row):
    if isinstance(row, list):
        return [str.lower(i.replace(" ", "")) for i in row]
    else:
        if isinstance(row, str):
            return str.lower(row.replace(" ", ""))
        else:
            return ""
```

```python
features = ["cast", "keywords", "genres"]
for feature in features:
    movies_new[feature] = movies_new[feature].apply(clean_data)
```

Now, let's whip up a "soup" combining all the metadata information extracted for input into the vectorizer.

```python
def create_soup(features):
    return ' '.join(features['keywords']) + ' ' + ' '.join(features['cast']) + ' ' + ' '.join(features['genres'])
```

```python
movies_new["soup"] = movies_new.apply(create_soup, axis=1)
print(movies_new["soup"].head())
```

```python
!pip install datasketch
from datasketch import MinHashLSHForest, MinHash
import pandas as pd
```

```python
forest = MinHashLSHForest(num_perm=128)
```

```python
for index, row in movies_new.iterrows():
    soup = create_soup(row)
    minhash = MinHash(num_perm=128)
    for word in soup.split():
        minhash.update(word.encode('utf-8'))
    forest.add(index, minhash)
```

##Get recommendations for the movies





```python
ratings_path = '/content/drive/MyDrive/DATASET/ratings.csv'
ratings_df = pd.read_csv(ratings_path)
```

```python
import random
def get_random_id(dataset):
  # Access the column of interest (assuming the column name is 'userId')
  column_data = dataset['userId']
  # Transform the column data to a list
  column_list = column_data.tolist()
  random_item = random.choice(column_list)
  return random_item



```

```python
def get_favorite_movies(ratings_df, movies_new, random_item):
  ratings_df = ratings_df[ratings_df['rating'] >= 4.0]
  # Filter the dataset based on the random user ID
  filtered_df = ratings_df[ratings_df['userId'] == random_item]
  # Extract the movie IDs from the filtered dataset
  filtered_ids = filtered_df['movieId'].tolist()
  # Assuming 'id' is the column name in 'movies_new' dataset containing the IDs
  # Assuming 'name' is the column name in 'movies_new' dataset containing the movie names

  # Create an empty list to store the movie names
  movie_names = []

  # Iterate over each ID in filtered_df
  for movie_id in filtered_df['movieId']:
      # Find the corresponding movie name in movies_new
      movie_name = movies_new.loc[movies_new['id'] == str(movie_id), 'title'].values
      # Check if a movie name was found
      if len(movie_name) > 0:
          # Append the movie name to the list
          movie_names.append(movie_name[0])
  return movie_names
```

```python
def get_recommendations(inputs):
  #Generate MinHash objects for the given movies:
  given_movies = [inputs[0], inputs[1], inputs[2]]
  given_minhashes = []
  for movie in given_movies:
      soup = create_soup(movies_new[movies_new['title'] == movie].iloc[0])
      minhash = MinHash(num_perm=128)
      for word in soup.split():
          minhash.update(word.encode('utf-8'))
      given_minhashes.append(minhash)
      #Query the LSH forest for similar items to the given movies
  similar_items = []
  k = 10  # Number of nearest neighbors to retrieve
  for minhash in given_minhashes:
      similar_items.extend(forest.query(minhash, k=k))
  #Count the occurrences of each similar item
  item_counts = {}
  for item in similar_items:
      if item not in item_counts:
          item_counts[item] = 0
      item_counts[item] += 1
  #Sort the items based on their counts in descending order
  sorted_items = sorted(item_counts.items(), key=lambda x: x[1], reverse=True)
  #Retrieve the top 10 recommended movies
  recommended_indices = [item[0] for item in sorted_items[:15]]
  recommended_movies = movies_new.iloc[recommended_indices]
  return recommended_movies
```

```python
flag = True
while(flag):
  x = get_random_id(ratings_df)
  print("Hello. Your id :", x)
  movie_names = get_favorite_movies(ratings_df, movies_new, x)
  print("This is your favorite movie list, please select three items between them :", movie_names)
  given = []
  for i in range (0,3):
    user_input = input("Please enter your input: ")
    given.append(user_input)
  outputs = get_recommendations(given)
  print("Here you go:")
  for index, movie in outputs.iterrows():
        print(movie['title'])
  print("connect again? y/n")
  user_input = input()
  if(user_input == 'y'):
    continue
  else:
    break
```
